```python
DATA_PATH = '/content/drive/MyDrive/프로젝트/뉴스 토픽/'
en_train = pd.read_csv(DATA_PATH + 'labelled_newscatcher_dataset.csv', sep = ';')
pol = pd.read_csv(DATA_PATH + 'politics.csv', encoding = 'cp949')
```

## 불필요한 Column 제거
concat을 위해 column 이름을 통일

```python
label0_5 = en_train.drop(['link', 'domain', 'published_date', 'lang'], axis = 1)
pol = pol.drop(['index'], axis = 1)
pol.rename(columns = {'topic_idx':'topic'}, inplace = True)
```

```python
pd.DataFrame(label0_5['title'].apply(len)).describe()
```

```python
pd.DataFrame(pol['title'].apply(len)).describe()
```

```python
label0_6 = pd.concat([label0_5, pol], axis = 0)
```

```python
label0_6.rename(columns = {'topic':'topic_idx'}, inplace = True)
```

```python
topic_dict = {'BUSINESS': 1,
 'ENTERTAINMENT': 3,
 'NATION': 2,
 'POLITICS': 6,
 'SCIENCE': 0,
 'SPORTS': 5,
 'TECHNOLOGY': 0,
 'WORLD': 4}
```

```python
def function(row):
    for topic, idx in topic_dict.items():
      if row == topic:
        return idx

label0_6['topic_idx'] = label0_6['topic_idx'].apply(function)
```

```python
label0_6.dropna(axis = 0, inplace = True)
```

```python
label0_6.head()
```

```python
label0_6['title_len'] = label0_6['title'].apply(len)
```

```python
pol['title_len'] = pol['title'].apply(len)
```

```python
df = label0_6[(100 >= label0_6['title_len']) & (label0_6['title_len'] >= 10)]
df.describe()
```
